```python
import numpy as np

customers = [150, 200, 180, 220 , 190, 250, 170, 1600, 210, 230]

customers_sorted = sorted(customers)

range_value = max(customers_sorted) - min(customers_sorted)
print(range_value)

left = customers_sorted[: len(customers_sorted) // 2]
right = customers_sorted[len(customers_sorted) // 2 :]

q1 = np.median(left)
q2 = np.median(customers_sorted)
q3 = np.median(right)
iqr = q3 - q1
print(f"Q1: {q1}\nQ2: {q2}\nQ3: {q3}\nIQR: {iqr}")

std_dev = np.std(customers_sorted)
print(std_dev)
```

Розрахуй діапазон (range) кількості відвідувачів.
Знайди квартилі (Q1, Q2, Q3) для кількості відвідувачів. Пам'ятай, що потрібно попередньо відсортувати вибірку.
Розрахуй міжквартильний діапазон (IQR).
Обчисли стандартне відхилення для кількості відвідувачів.

На основі отриманих результатів проаналізуй:
наскільки стабільними були відвідування вебсайту протягом цього періоду;
які типові значення кількості відвідувачів;
чи були екстремальні значення.

Судячи з квартилів відвідування сайту було достатньо стабільним.
Типові значення в районі 150 - 250.
Діапазон та стандартнє відхилення дає нам розуміння що є одне екстримальне значення, і судячи з квартилів це максимальне значення.
